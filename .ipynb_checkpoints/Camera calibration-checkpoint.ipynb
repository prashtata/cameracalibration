```python
import cv2
import numpy as np
import os
import glob
```

```python
CHECKERBOARD = (10,11)
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)

objpoints = []
imgpoints = [] 

objp = np.zeros((1, CHECKERBOARD[0] * CHECKERBOARD[1], 3), np.float32)
objp[0,:,:2] = np.mgrid[0:CHECKERBOARD[0], 0:CHECKERBOARD[1]].T.reshape(-1, 2)
prev_img_shape = None
```

```python
images = glob.glob('./cropped_images/*.png')

for fname in images:
    img = cv2.imread(fname)
    img = cv2.resize(img, (0, 0), fx = 5.0, fy = 5.0)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    
    # Color-segmentation to get binary mask
    lwr = np.array([0, 0, 143])
    upr = np.array([179, 61, 252])
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    msk = cv2.inRange(hsv, lwr, upr)

    # Extract chess-board
    krn = cv2.getStructuringElement(cv2.MORPH_RECT, (50, 30))
    dlt = cv2.dilate(msk, krn, iterations=5)
    res = 255 - cv2.bitwise_and(dlt, msk)

    # Displaying chess-board features
    res = np.uint8(res)
    
#     ret, corners = cv2.findChessboardCorners(res, CHECKERBOARD, flags = cv2.CALIB_CB_ADAPTIVE_THRESH + cv2.CALIB_CB_FAST_CHECK + cv2.CALIB_CB_NORMALIZE_IMAGE)
#     # If desired number of corners are found in the image then ret = true
#     print(ret)
#     if ret == True:
#         objpoints.append(objp)
#         print("ok1")
#         corners2 = cv2.cornerSubPix(gray, corners, (11,11),(-1,-1), criteria)
#         print("ok2")
#         imgpoints.append(corners2)
#         print(corners2)
#         img = cv2.drawChessboardCorners(img, CHECKERBOARD, corners2, ret)
     
#     cv2.imshow('img',img)
#     cv2.waitKey(0)
 
# cv2.destroyAllWindows()
 
# h,w = img.shape[:2]

ret, corners = cv.findChessboardCorners(res, CHECKERBOARD, 
                                                flags = cv.CALIB_CB_ADAPTIVE_THRESH +
                                                        cv.CALIB_CB_FAST_CHECK + 
                                                        cv.CALIB_CB_NORMALIZE_IMAGE)

        print('ret=',ret)
        # image.show()     


# If desired number of corners are detected, we refine the pixel coordinates and display them on the images of checker board
        
    if ret == True:
        # print("true")
        objpoints.append(objp)
        # refining pixel coordinates for given 2d points.
        corners2 = cv.cornerSubPix(gray,corners,(11,11),(-1,-1),criteria)
        
        imgpoints.append(corners2)

        # Draw and display the corners
        img = cv.drawChessboardCorners(img, CHECKERBOARD, corners2,ret)
        # print(f'Loop Number{i}')
        status2 = cv.imwrite('/C:/Users/AMIT KUMAR/crop_image/detected_image'+str(p)+'.jpg',img)
        print('image saved to folder', status2)
        cv.imshow('image'+str(p),img)

            #writing the images in a folder output_images
            # cv.imwrite('/Users/surajkumar/Documents/TIFR/veinimaging/calib_recon3x3/cropped_images/image{}.png'.format(bb), img)

        cv.waitKey()
        cv.destroyAllWindows()
        h,w = img.shape[:2]
        # print(h,w)

# #  Performing camera calibration by  passing the value
#  of known 3D points (objpoints) and corresponding 
# pixel coordinates of the  detected corners (imgpoints)
        ret, mtx, dist, rvecs, tvecs = cv.calibrateCamera(objpoints, imgpoints, gray.shape[::-1],None,None)

#         # rvecs = rotational vector
#         # tvecs = translation vector
#         # dist  = distortion coefficient
#         # camera matrix = Intrinsic camera parameters  

#         # print("Camera matrix : \n") 
        print(mtx)
        print("dist : \n")
        print(dist)
        print("rvecs : \n")
        print(rvecs)
        print("tvecs : \n")
        print(tvecs)
        print("Calibration done using image",p)
#         p+=1
    else:
        print("Desired number of corners are not detected for Image ", p)
#         p+=1
```

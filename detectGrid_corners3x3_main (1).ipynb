```python
g1 = np.copy(gp)
max = 2000
for i in range(0,gs):  
    for j in range(0,gs):
        # print('j=',j)
        m = max
        u = i*gs
        v = u + gs
        
        while u < v:
            # print(u)
            if g1[u,1] < m:
                m = g1[u,1]
                indu = u
            u = u + 1
            grid[0,i,j] = g1[indu,0]
            grid[1,i,j] = g1[indu,1]
        g1[indu,1] = max
# print(grid) 
intgrid= grid.astype(int)
print(intgrid)
# print(gp) 
# print(grid[:,0,0])
```

```python
# Defining the dimensions of checkerboard
CHECKERBOARD = (10,11)
criteria = (cv.TERM_CRITERIA_EPS + cv.TERM_CRITERIA_MAX_ITER, 30, 0.001)
world_scaling = 1
# Creating vector to store vectors of 3D points for each checkerboard image
objpoints = [] 
# Creating vector to store vectors of 2D points for each checkerboard image
imgpoints = [] 

# Defining the world coordinates for 3D points
points = objp = np.zeros((CHECKERBOARD[0]*CHECKERBOARD[1], 3), np.float32)
p = objp[:,:2] = np.mgrid[0:CHECKERBOARD[0], 0:CHECKERBOARD[1]].T.reshape(-1, 2)
objp = world_scaling*objp
# print(p)
p = 0
# prev_img_shape = None 
for i in range(0,3):
    for j in range(0,3): 
        image_arr = grey[ygrid[j,i]:ygrid[j+1,i+1],xgrid[j,i]:xgrid[j+1,i+1]]
        image = Image.fromarray(image_arr)
        # image.show() 
        status1 = cv.imwrite('./cropped_images/crop_image'+str(p)+'.jpg',image_arr)
        # print('image is saved', status)
        # print('p=', p)
        
        images = './cropped_images/crop_image'+str(p)+'.jpg'
        img = cv.imread(images) 
        gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)

        ret, corners = cv.findChessboardCorners(gray, CHECKERBOARD, 
                                                flags = cv.CALIB_CB_ADAPTIVE_THRESH +
                                                        cv.CALIB_CB_FAST_CHECK + 
                                                        cv.CALIB_CB_NORMALIZE_IMAGE)

        print('ret=',ret)
        # image.show()     


# If desired number of corners are detected, we refine the pixel coordinates and display them on the images of checker board
        
        if ret == True:
            # print("true")
            objpoints.append(objp)
            # refining pixel coordinates for given 2d points.
            corners2 = cv.cornerSubPix(gray,corners,(11,11),(-1,-1),criteria)
            
            imgpoints.append(corners2)

            # Draw and display the corners
            img = cv.drawChessboardCorners(img, CHECKERBOARD, corners2,ret)
            # print(f'Loop Number{i}')
            status2 = cv.imwrite('./det_images/detected_image'+str(p)+'.jpg',img)
            print('image saved to folder', status2)
            cv.imshow('image'+str(p),img)

            #writing the images in a folder output_images
            # cv.imwrite('/Users/surajkumar/Documents/TIFR/veinimaging/calib_recon3x3/cropped_images/image{}.png'.format(bb), img)

            cv.waitKey()
            cv.destroyAllWindows()
            h,w = img.shape[:2]
            # print(h,w)

# #  Performing camera calibration by  passing the value
#  of known 3D points (objpoints) and corresponding 
# pixel coordinates of the  detected corners (imgpoints)
            ret, mtx, dist, rvecs, tvecs = cv.calibrateCamera(objpoints, imgpoints, gray.shape[::-1],None,None)

#         # rvecs = rotational vector
#         # tvecs = translation vector
#         # dist  = distortion coefficient
#         # camera matrix = Intrinsic camera parameters  

#         # print("Camera matrix : \n") 
            print(mtx)
            print("dist : \n")
            print(dist)
            print("rvecs : \n")
            print(rvecs)
            print("tvecs : \n")
            print(tvecs)
            print("Calibration done using image",p)
            p+=1
        else:
            print("Desired number of corners are not detected for Image ", p)
            p+=1


# Note: When performed on the 3x3 checker board, image 6 is not recognised due to image quality. 
# Hence, you see only 8 of 9 detected corner images
```
